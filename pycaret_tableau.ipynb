```python
import pandas as pd
```

```python
dataset = pd.read_csv('data/direct_marketing.csv')
```

```python
data = dataset.sample(frac=0.95, random_state=786).reset_index(drop=True)
data_unseen = dataset.drop(data.index).reset_index(drop=True)
new_customers = data_unseen.drop(['visit'], axis=1)
new_customers.to_csv('data/new_customers.csv', index=False, header=True)
```

```python
print('Data for Modeling: ' +str(data.shape))
print('Unseen Data for Predictions: ' +str(data_unseen.shape))
```

```python
from pycaret.classification import *
```

```python
clf = setup(data, target='visit', session_id=562, ignore_features=['ID'], silent=True)
```

```python
nb = create_model('nb')
```

```python
nb_final = finalize_model(nb)
```

```python
save_model(nb_final, 'final_models/nb_direct')
```
